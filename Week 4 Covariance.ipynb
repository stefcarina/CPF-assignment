```python
## risk is not abstract. 
## each state produces a payoff pair. each state is a point in the payoff plane. they can forms a cloud of points.
## they have joint risk, which means the two assets move together across states.
## the covaraiance matrix has individual risk as well as joint risk.
## which means the covariance matrix considers all state- by- state joint situation 
```

```python
## from the formular, the correlation is covariance / times of standard deviation of asset 1,and 2
## covariance has units. if it rescale, it will change
## correlation is units free.and has a boundary of -1,1
```

```python
## based on the formular, variance depends on the weights.
##increase the weights will increase variance.
## covariance determines from the joints risk of two asset.
```

```python
## eigenvalues and eigenvectors describe the shape, size and length of the risk.
## they can visualize the covariance matrix.
## eigenvectors identified the direction of joint risk.
## eigenvalues identified the scales / size / shape
```

```python
## the minmium-variance portfolio has the smallest eigenvalue
## thus the ellipse is the narrowest
```

```python
import numpy as np

def port(w,sigma1, sigma2, rho):
    Sigma = np.array((
        (sigma1**2, rho*sigma1*sigma2),
        (rho*sigma1*sigma2, sigma2 **2)
    ))
    weights = np.array((w, 1-w))

    return weights @ Sigma @ weights

sigma1 = 0.2
sigma2 = 0.1
rho = 0.5
w = 0.6

var_p = port(w,sigma1, sigma2, rho)
var_p
```

```python
import numpy as np

def port_var(w,sigma1, sigma2, rho):
    Sigma = np.array((
        (sigma1**2, rho*sigma1*sigma2),
        (rho*sigma1*sigma2, sigma2 **2)
    ))
    weights = np.array((w, 1-w))

    return weights @ Sigma @ weights

def w_minvar(sigma1, sigma2, rho):
    num = sigma2 **2 - rho* sigma1 * sigma2
    de = sigma1**2 + sigma2 **2 - 2*rho*sigma1*sigma2
    return num / de


sigma1 = 0.5
sigma2 = 0.1
rho = 0.5

w_star = w_minvar(sigma1, sigma2, rho)
var = port_var(w_star,sigma1,sigma2,rho)    
w_star , var


## when rho increase, diversification week
```
