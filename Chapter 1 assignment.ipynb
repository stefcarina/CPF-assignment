## Given two traded assets with payoffs B1 and S1 and prices B0 and S0, compute the payoff of a portfolio that costs exactly 1 at t = 0

```python
import numpy as np
def re(B0, B1, S0, S1, w):
    return w *(B1 - B0) / B0 + (1-w) * (S1 -S0)/ S0

re(10,20,5,4,0.4)
```

## choose a payoff target C1 and solve M*phi = C1 when M = [B1 S1 ] is a square; verify by checking M* phi -C1 = 0

```python
B = (100, np.array((110, 110)))
S = (100, np.array((80,120)))

M = np.array((B[1], S[1])).T

K = 110
C1 = np.maximum(S[1]- K, 0)
k2 = 118
C2 = np.maximum(S[1]- k2, 0)

phi = np.linalg.solve(M, C1)
phi2 = np.linalg.solve(M, C2)
phi, phi2

# test
np.allclose(np.dot(M, phi), C1)
```
