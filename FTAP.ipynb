```python
import numpy as np

def is_complete_market(S1, tol=1e-12):
    S1 = np.asarray(S1, dtype=float)
    m = S1.shape[0]              # number of states (should be 3 here)
    rank = np.linalg.matrix_rank(S1, tol=tol)
    return (rank == m), rank

# Example usage:
S1 = np.array([
    [1.05, 140],
    [1.05,  90],
    [1.05,  60],
], dtype=float)

complete, rank = is_complete_market(S1)
complete,rank
```

```python
import numpy as np

r = 0.05

B0 = 1.0
B1 = np.array([1+r, 1+r])     

S0_stock = 100.0
Su, Sd = 130.0, 80.0
S1_stock = np.array([Su, Sd])    

S0 = np.array([B0, S0_stock])                     
S1 = np.column_stack([B1, S1_stock])                


pi = np.linalg.solve(S1.T, S0)             


K = 100.0
X = np.maximum(K - S1_stock, 0.0)            

price_state_prices = pi @ X

theta = np.linalg.solve(S1, X)            
b, Delta = theta

price_replication = theta @ S0


print("State prices π =", pi)
print("Price via state prices =", price_state_prices)
print("Replicating portfolio (b, Delta) =", (b, Delta))
print("Price via replication =", price_replication)
print("Difference =", price_state_prices - price_replication)
```

```python
import numpy as np

S0 = np.array([1.0, 0.4])

S1 = np.array([
    [1.0, 2.0],
    [1.0, 0.5]
])

theta = np.array([-0.4, 1.0])

V0 = theta @ S0
V1 = S1 @ theta

is_arbitrage = (V0 <= 0) and np.all(V1 >= 0) and np.any(V1 > 0)

print("V0 =", V0)
print("V1 =", V1)
print("Arbitrage", is_arbitrage)
```
